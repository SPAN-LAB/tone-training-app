```python
import os
import pandas as pd
import re
path = "/Volumes/gurindapalli/projects/Plasticity_training/tone-training-app/participants/livia/Perception with Full Feedback/session_tracking/session5.csv"
```

```python
session_tracking_folder = os.path.dirname(path)
```

```python
session_tracking_folder
```

```python
files = os.listdir(session_tracking_folder)
for file in files:
    print(int(re.findall(r'\d+', file)[0]))
```

```python
content = []
for file in os.listdir(session_tracking_folder):
    if file.endswith("csv"):
        path = os.path.join(session_tracking_folder, file)
        df_temp = pd.read_csv(path)
        df_temp["session"] = int(re.findall(r'\d+', file)[0])
        content.append(df_temp)

df = pd.concat(content, axis=0, ignore_index=True)
df["accuracy"] = (df["accuracy"].fillna(0) * 100).round(2)
df
```
